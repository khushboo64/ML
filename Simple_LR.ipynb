```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('Simple_LR.csv')
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
